#### There are strings in genre col, not lists, so in the future we will work on it accordingly

```python
import ast


def parse_genres(val):
    if pd.isna(val):
        return []
    if isinstance(val, list):
        return val
    s = str(val).strip()
    # try literal_eval for strings like "['Comedy', 'Fantasy']"
    try:
        parsed = ast.literal_eval(s)
        return parsed if isinstance(parsed, (list, tuple)) else [parsed]
    except Exception:
        # fallback: remove surrounding brackets and split by comma
        s2 = s.strip("[]")
        parts = [p.strip().strip('\"\'') for p in s2.split(',') if p.strip()]
        return parts


# create a new column with lists
df_animes_head['genres_list'] = df_animes_head['genre'].apply(parse_genres)

# list with all unique genres (sorted)
all_genres = sorted({g for lst in df_animes_head['genres_list'] for g in lst})

# optional: counts per genre
genre_counts = pd.Series([g for lst in df_animes_head['genres_list'] for g in lst]).value_counts()

print(all_genres)
print(f'genre counts:\n{genre_counts}')
```

#### Created the new col with list values for genres

```python
import re


def _col_name(genre):
    s = re.sub(r'[^0-9A-Za-z]+', '_', genre).strip('_').lower()
    return f'genre_{s}'


# ensure `genres_list` exists (recompute if needed)
if 'genres_list' not in df_animes_head.columns:
    df_animes_head['genres_list'] = df_animes_head['genre'].apply(parse_genres)

# recompute unique genres if `all_genres` not present
all_genres = sorted({g for lst in df_animes_head['genres_list'] for g in lst})

for g in all_genres:
    col = _col_name(g)
    df_animes_head[col] = df_animes_head['genres_list'].apply(lambda lst: int(g in lst))

# optional: verify
print(sorted([c for c in df_animes_head.columns if c.startswith('genre_')])[:20])
```

#### Parsed all genres, did not miss any

```python
df_animes_head = df_animes_head.drop(columns=['genre'])
df_animes_head.head(10)
```

```python
df_animes_head.isnull().sum()
```

```python
df_animes_head['synopsis'] = df_animes_head['synopsis'].fillna('no synopsis provided')
df_animes_head.isnull().sum()
```

```python
genre_cols = [c for c in df_animes_head.columns if isinstance(c, str) and c.startswith('genre_')]

genre_means = {}
for col in genre_cols:
    mean_val = df_animes_head.loc[df_animes_head[col] == 1, 'score'].mean()
    genre_means[col] = mean_val

global_mean = df_animes_head['score'].mean()


def _fill_score_for_row(row):
    if pd.notna(row['score']):
        return row['score']
    # find genres present for this anime
    present = [g for g in genre_cols if row.get(g) == 1]
    # collect non-null genre means
    vals = [genre_means[g] for g in present if pd.notna(genre_means.get(g))]
    if vals:
        return float(sum(vals) / len(vals))
    return float(global_mean)


df_animes_head['score'] = df_animes_head.apply(_fill_score_for_row, axis=1)

df_animes_head['score'] = df_animes_head['score'].round(2)

print('Missing scores after fill:', df_animes_head['score'].isnull().sum())
```

```python

df_animes_head['episodes'] = pd.to_numeric(df_animes_head['episodes'], errors='coerce')

# Recompute genre indicator columns
genre_cols = [c for c in df_animes_head.columns if isinstance(c, str) and c.startswith('genre_')]

genre_medians = {}
for col in genre_cols:
    med = df_animes_head.loc[df_animes_head[col] == 1, 'episodes'].median()
    genre_medians[col] = med

# Global median
global_median = df_animes_head['episodes'].median()

# Fallback if global_median is NaN
if pd.isna(global_median):
    global_median = 1


# Function to fill a single row's episodes
def _fill_episodes_for_row(row):
    if pd.notna(row['episodes']):
        return row['episodes']
    # if anime is a movie, set episodes to 1
    if 'type' in row.index and isinstance(row['type'], str) and row['type'].strip().lower() == 'movie':
        return 1
    # find genres present for this anime
    present = [g for g in genre_cols if row.get(g) == 1]
    # collect non-null genre medians
    vals = [genre_medians[g] for g in present if pd.notna(genre_medians.get(g))]
    if vals:
        # use median of genre medians to be robust
        return float(np.median(vals))
    return float(global_median)


# Apply filling
df_animes_head['episodes'] = df_animes_head.apply(_fill_episodes_for_row, axis=1)

df_animes_head['episodes'] = df_animes_head['episodes'].round().astype(int)

print('Missing episodes after fill:', df_animes_head['episodes'].isnull().sum())
```

```python
df_animes_tale['episodes'] = pd.to_numeric(df_animes_tale['episodes'], errors='coerce')

genre_cols = [c for c in df_animes_tale.columns if isinstance(c, str) and c.startswith('genre_')]

genre_medians_tale = {}
for col in genre_cols:
    med = df_animes_tale.loc[df_animes_tale[col] == 1, 'episodes'].median()
    genre_medians_tale[col] = med

if 'type' in df_animes_tale.columns:
    global_median_tale = df_animes_tale.loc[df_animes_tale['type'] != 'Movie', 'episodes'].median()
else:
    global_median_tale = df_animes_tale['episodes'].median()

if pd.isna(global_median_tale):
    global_median_tale = 1


# Row-wise filler
def _fill_episodes_tale_row(row):
    if pd.notna(row['episodes']):
        return row['episodes']
    if 'type' in row.index and isinstance(row['type'], str) and row['type'].strip().lower() == 'movie':
        return 1
    present = [g for g in genre_cols if row.get(g) == 1]
    vals = [genre_medians_tale[g] for g in present if pd.notna(genre_medians_tale.get(g))]
    if vals:
        return float(np.median(vals))
    return float(global_median_tale)


df_animes_tale['episodes'] = df_animes_tale.apply(_fill_episodes_tale_row, axis=1)
df_animes_tale['episodes'] = df_animes_tale['episodes'].round().astype(int)

print('Missing episodes in tale after fill:', df_animes_tale['episodes'].isnull().sum())

```
