```python
df_profile_uid = df_reviews[["profile", "uid"]].drop_duplicates("profile")

df_profiles_clean = (
    df_profiles
    .merge(df_profile_uid, on="profile", how="inner")
    .drop(columns=["profile"])  # убираем грязные ники
)

df_profiles_clean.head(10)
```

```python
df_reviews = df_reviews.drop(columns=['profile'])
df_reviews.head(10)
```

```python
df_reviews.to_parquet("../../data/processed/reviews_final_cleaned.parquet")
df_profiles_clean.to_parquet('../../data/processed/profiles_cleaned_final.parquet')
```
